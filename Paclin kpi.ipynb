```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kpi logistica.csv',delimiter=';')
```

```python
df.info()
```

```python
df.head()
```

```python
plt.figure(figsize=(9,5),dpi=180)
sns.scatterplot(data=df,x='Mes y Año',y='Fill Rate')
```

```python
plt.figure(figsize=(12,7),dpi=180)
sns.scatterplot(data=df,x='SKU',y='Fill Rate')

# SKU request vs Fill Rate
```

```python
plt.figure(figsize=(12,7),dpi=180)
sns.scatterplot(data=df,x='SKU',y='Fill Rate',hue='Entregado')

# SKU request vs Fill Rate, filtering 'Entregados'
```

```python
df.nlargest(10,'SKU')

# top 10 request´s with largest SKU
```

```python
plt.figure(figsize=(12,7),dpi=180)
sns.barplot(data=df, x='Preparo',y='Entregado',order=('Luraschi I','Saravia F','Sanchez G','Chicco L','Paez F','Diaz D'))

# Plot full preparation vs Person
```

```python
plt.figure(figsize=(12,7),dpi=180)
sns.barplot(data=df, x='Preparo',y='Entregado',hue='Mes y Año',ci=None,palette='bright',order=('Luraschi I','Saravia F','Sanchez G','Chicco L','Paez F','Diaz D'))

# Person vs SKU delivered , by Month/year
```

```python
plt.figure(figsize=(11,6),dpi=180)
sns.countplot(data=df,x='Preparo',hue='Mes y Año',order=('Luraschi I','Saravia F','Sanchez G','Chicco L','Paez F','Diaz D'))

plt.legend(loc=(.87,.3))

# Person vs Count of request, by month/year
```

```python
plt.figure(figsize=(12,7),dpi=180)
sns.countplot(data=df,hue='Preparo',x='Reclamo')

plt.legend(loc=(.7,.5))
```

```python
plt.figure(figsize=(12,7),dpi=180)
sns.countplot(data=df,x='Reclamo',hue='Controlo')

plt.legend(loc=(.7,.5))
```

```python
plt.figure(figsize=(12,7),dpi=180)
sns.countplot(data=df,x='Reclamo',hue='Responsable',saturation=.9,)

plt.legend(loc=(.7,.5))
```

```python
plt.figure(figsize=(10,10),dpi=250)
sns.catplot(data=df,x='Preparo',y='SKU',hue='Area Prep',col='Controlo')
```
